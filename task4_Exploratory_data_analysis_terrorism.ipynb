### Extracting required columns

```python
data=data[['iyear','imonth','iday','country_txt','provstate','region_txt','city','latitude','longitude','attacktype1_txt',
           'nkillu','nwound','targtype1_txt','summary','gname','weaptype1_txt','motive',]]
```

```python
data
```

```python
data.isnull().sum()
```

As seen, provstate,city,latitude,longitude,summary,motive ,nkillu and nwound have null values. Hence we assign 0 to the null rows and group nkillu and nwound under casualities

```python
data['nkillu']=data['nkillu'].fillna(0)
data['nwound']=data['nwound'].fillna(0)
data['Casualities']=data['nkillu']+data['nwound']
```

### Visual Representation

```python
year=data['iyear'].unique()
year_c=data['iyear'].value_counts(dropna=False).sort_index()
plt.figure(figsize=(20,10))
sb.barplot(x=year,y=year_c,palette='Dark2_r')
plt.xticks(rotation=50)
plt.xlabel('YEAR',fontsize=15)
plt.ylabel('NUMBER OF ATTACKS',fontsize=15)
plt.title('YEAR vs NUMBER OF ATTACKS IN EACH YEAR',fontsize=15)
plt.show()
```

```python
data['iyear'].value_counts()
```

```python
plt.rcParams["figure.figsize"] = 20,10
sb.barplot(x = data["country_txt"].value_counts()[:20].index, y=data["country_txt"].value_counts()[:20].values,palette='Greens_r')
plt.xticks(rotation=50)
plt.xlabel("Country", size=15)
plt.ylabel("Number", size=15)
plt.title("COUNTRY WISE ATTACKS", size=15)
```

```python
df=data['country_txt'].value_counts()
print('Top 5:')
print(df.head(5))
print('\nLast 5:')
print(df.tail(5))
```

```python
plt.rcParams["figure.figsize"] = 20,10
sb.barplot(x = data["city"].value_counts()[:20].index, y=data["city"].value_counts()[:20].values,palette='GnBu_r')
plt.xticks(rotation=50)
plt.xlabel("City", size=15)
plt.ylabel("Number", size=15)
plt.title("CITY WISE ATTACKS", size=15)
```

```python
df = data[['iyear','nkillu']].groupby(['iyear']).sum()
fig, ax4 = plt.subplots(figsize=(20,10))
df.plot(kind='bar',alpha=0.7,ax=ax4)
plt.xticks(rotation = 50)
plt.title("NUMBER OF DEATHS",fontsize=15)
plt.ylabel("NUMBER OF PEOPLE KILLED",fontsize=15)
plt.xlabel('YEAR',fontsize=15)
top_side = ax4.spines["top"]
top_side.set_visible(False)
right_side = ax4.spines["right"]
right_side.set_visible(False)
```

```python
c=['darkblue','blue','skyblue','lightblue','pink','violet','magenta','purple','darkviolet']
data['attacktype1_txt'].value_counts().plot(kind='bar',figsize=(20,10),color=c)
plt.xticks(rotation = 70)
plt.xlabel("TYPE OF ATTACKS",fontsize=15)
plt.ylabel("NUMBER OF ATTACKS",fontsize=15)
plt.title("NUMBER OF ATTACKS BASED ON TYPE OF ATTACKS",fontsize=15)
plt.show()
```

```python
data['attacktype1_txt'].value_counts()
```

```python
data[['attacktype1_txt','nkillu']].groupby(data['attacktype1_txt'],axis=0).sum().plot(kind='bar',figsize=(20,10),color='olive')
plt.xticks(rotation = 70)
plt.xlabel("TYPE OF ATTACKS",fontsize=15)
plt.ylabel("NUMBER OF KILLS",fontsize=15)
plt.title("NUMBER OF KILLS BASED ON TYPE OF ATTACKS",fontsize=15)
plt.show()
```

```python
data[['attacktype1_txt','nwound']].groupby(data['attacktype1_txt'],axis=0).sum().plot(kind='bar',figsize=(20,10),color='firebrick')
plt.xticks(rotation = 70)
plt.xlabel("TYPE OF ATTACKS",fontsize=15)
plt.ylabel("NUMBER OF WOUNDED",fontsize=15)
plt.title("NUMBER OF WOUNDED BASED ON TYPE OF ATTACKS",fontsize=15)
plt.show()
```

```python
plt.rcParams["figure.figsize"] = 20,10
sb.barplot(x = data["targtype1_txt"].value_counts()[:20].index, y=data["targtype1_txt"].value_counts()[:20].values,palette='PRGn_r')
plt.xticks(rotation=80)
plt.xlabel("TARGET TYPE", size=15)
plt.ylabel("NUMBER OF ATTACKS", size=15)
plt.title("ATTACKS BASED ON TARGETS", size=15)
```
